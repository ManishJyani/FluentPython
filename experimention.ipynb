```python
# This code is just to get visualization of first two stages 
# building AST and then compiling to Bytecode which interpreter will excute later stage
import ast, dis

code = "a = 2 + 3"

print("=== AST ===")
print(ast.dump(ast.parse(code), indent=4))

print("\n=== BYTECODE ===")
dis.dis(code)
```
