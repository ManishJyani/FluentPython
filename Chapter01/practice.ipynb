First Concept : Public vs Private  
In first chapter of fluent python , In FrenchDeck class I saw object attribute "self._cards", that made me curious about what is this underscore even mean?
From there I got to know the concept of public vs private attributes and methods.

```python
"""
Level 1 
self._cards = []
Meaning: "This is internal / private by convention"
Actually private? No — fully accessible from outside
Who respects it? Any developer who follows Python conventions
Extra behavior? One thing — if you do from module import *, names starting with _ are not imported
"""
```

```python
"""
Level 2 — Double Underscore __name (Name Mangling)
self.__cards = []
Meaning: "I really don't want you touching this"
Actually private? Still no — but Python mangles the name
What is name mangling? Python secretly renames __cards to _FrenchDeck__cards
So
deck.__cards      # AttributeError — name doesn't exist!
deck._FrenchDeck__cards  # Works — but now you're really forcing it
"""
```

```python
"""
Level 3 — Dunder __name__ (Double on Both Sides)
def __len__(self):
def __init__(self):
Meaning: These are Python's special methods — part of the data model
Not private at all — these are meant to be used by Python itself
Never name your own attributes like this — these are reserved for Python
"""
```

```python
"""
Why Does Python Do It This Way?
This is a deliberate philosophical choice. Python's philosophy is:

"We are all consenting adults here."

Python trusts developers. It doesn't lock things down. Instead it uses conventions and says:

If you see _name — respect it as private
If you really need to access it — you can, but you're on your own
If something breaks — that's on you

This is very different from Java's approach where the compiler enforces access control.
"""
```

```python
# Real word example
class BankAccount:
    
    def __init__(self, owner, balance):
        self.owner = owner           # Public attribute
        self._balance = balance      # Private attribute
        self.__transaction_key = 'X7K9'  # Strongly private

    # Public method — the interface you expose
    def deposit(self, amount):
        if self._is_valid_amount(amount):  # calls private method
            self._balance += amount
            self._log_transaction('deposit', amount)  # calls private method

    # Public method
    def withdraw(self, amount):
        if self._is_valid_amount(amount) and self._has_sufficient_funds(amount):
            self._balance -= amount
            self._log_transaction('withdrawal', amount)

    # Public method
    def get_balance(self):
        return self._balance

    # Private method — internal validation helper
    def _is_valid_amount(self, amount):
        return amount > 0

    # Private method — internal check
    def _has_sufficient_funds(self, amount):
        return self._balance >= amount

    # Private method — internal logging
    def _log_transaction(self, type, amount):
        print(f"Transaction: {type} of {amount}")
```
